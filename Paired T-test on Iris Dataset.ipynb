```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load dataset
iris_data = load_iris(as_frame=True)
df = iris_data.frame
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
print("Missing values in each column:")
print(df.isnull().sum(), "\n")

# -----------------------
# 4. Distribution of species
# -----------------------
print("Species count:")
print(df['species'].value_counts(), "\n")

plt.figure(figsize=(6,4))
sns.countplot(x='species', data=df)
plt.title("Count of Each Species")
plt.show()
```

```python
# 5. Pairplot to see relationships between features
# -----------------------
sns.pairplot(df, hue='species', diag_kind='kde')
plt.suptitle("Pairwise Relationships between Features", y=1.02)
plt.show()
```

```python
# 6. Correlation heatmap
# -----------------------
plt.figure(figsize=(6,4))
sns.heatmap(df.drop(columns='species').corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# 7. Boxplots to visualize spread of each feature by species
# -----------------------
plt.figure(figsize=(12,8))
for i, column in enumerate(df.columns[:-2]):  # skip target & species
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='species', y=column, data=df)
    plt.title(f"{column} by Species")
plt.tight_layout()
plt.show()
```

```python
#Part 1: One-sample t-test
import scipy.stats as stats # Import stats module
alpha = 0.05 # Define significance level
setosa_sepal_length = df[df['species'] == 'setosa']['sepal length (cm)']
t_stat, p_value = stats.ttest_1samp(setosa_sepal_length, 5.0)
print("\nOne-sample t-test:")
print(f"Mean: {setosa_sepal_length.mean():.3f}, T-stat: {t_stat:.4f}, P-value: {p_value:.6f}")
if p_value < alpha:
 print("Reject H0")
else:
 print("Fail to Reject H0")
```

```python
import scipy.stats as stats

alpha = 0.05
setosa_sepal_length = df[df['species'] == 'setosa']['sepal length (cm)']
versicolor_sepal_length = df[df['species'] == 'versicolor']['sepal length (cm)']

t_stat, p_value = stats.ttest_ind(setosa_sepal_length, versicolor_sepal_length)
print("\nTwo-sample t-test (independent):")
print(f"Setosa mean: {setosa_sepal_length.mean():.3f}, Versicolor mean: {versicolor_sepal_length.mean():.3f}")
print(f"T-stat: {t_stat:.4f}, P-value: {p_value:.6f}")

if p_value < alpha:
    print("Reject H0: Means are significantly different")
else:
    print("Fail to Reject H0: No significant difference")
```

```python
import scipy.stats as stats

alpha = 0.05
setosa_data = df[df['species'] == 'setosa']
sepal_length = setosa_data['sepal length (cm)']
sepal_width = setosa_data['sepal width (cm)']

t_stat, p_value = stats.ttest_rel(sepal_length, sepal_width)
print("\nPaired t-test:")
print(f"Mean sepal length: {sepal_length.mean():.3f}, Mean sepal width: {sepal_width.mean():.3f}")
print(f"T-stat: {t_stat:.4f}, P-value: {p_value:.6f}")

if p_value < alpha:
    print("Reject H0: Means are significantly different")
else:
    print("Fail to Reject H0: No significant difference")
```

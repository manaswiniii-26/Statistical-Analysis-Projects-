```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up seaborn for better visuals
sns.set(style="whitegrid")

# Generating sample data
np.random.seed(0)
data1 = np.random.normal(50, 10, 30)  # Sample 1
data2 = np.random.normal(55, 10, 30)  # Sample 2
data3 = data1 + np.random.normal(0, 5, 30)  # Paired sample to data1

print("data1\n", data1)
print("\ndata2\n", data2)
print("\ndata3\n", data3)


def one_sample_z_test_and_plot(sample, mu, sigma, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_stat = (sample_mean - mu) / (sigma / np.sqrt(n))
    p_val = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    print("One-Sample z-Test")
    print(f"Sample Mean: {sample_mean}")
    print(f"Z-Statistic: {z_stat}")
    print(f"P-Value: {p_val}")

    # Plotting
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)
    critical_z = stats.norm.ppf(1 - alpha / 2)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution')
    plt.axvline(x=z_stat, color='r', linestyle='--', label='z-statistic')
    plt.axvline(x=critical_z, color='b', linestyle='--', label='Critical value (+)')
    plt.axvline(x=-critical_z, color='g', linestyle='--', label='Critical value (-)')
    plt.fill_between(x, y, where=abs(x) >= critical_z, color='gray', alpha=0.5)
    plt.legend()
    plt.title('One-Sample z-Test')
    plt.xlabel('z value')
    plt.ylabel('Probability Density')
    plt.show()

    if p_val < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")


# Assuming we know the population standard deviation
mu = 50  # Hypothesized population mean
sigma = 10
one_sample_z_test_and_plot(data1, mu, sigma, alpha=0.05)
```
